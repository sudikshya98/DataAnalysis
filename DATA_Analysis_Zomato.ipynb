```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create the dataframe
dataframe = pd.read_csv('/content/Zomatodata.csv')
print(dataframe)
```

```python
dataframe
```

```python
dataframe.info
```

```python
#Convert the data type of the rate column
def handleRate(value):
  value = str(value).split('/')[0]
  value = value.replace(',','')
  return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info
```

```python
#Type of Restaurant
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type of Restaurant')
```

Conclusion - There are a majority of Dining Restaurants registered on Zomato

```python
# Visualizing the type of restaurants with their votes
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="red", marker = "o")
plt.xlabel("Type of Restaurant", c="black", size=10)
plt.ylabel("Votes", c="black", size=10)
```

Conclusion - Dining Restaurants has received maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'], bins=20)
plt.title("Ratings Distribution")
plt.show()
```

Conclusions - A majority of restaurants are receiving an average rating of 3.75-4.00

```python
#Observing which mode has better rating
dataframe.head()
```

```python
plt.figure(figsize = (3,6))
sns.barplot(x = 'online_order', y = 'rate', data = dataframe)
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt= 'd')
plt.title("HeatMap")
plt.xlabel("Online Order")
plt.ylabel("listed_in(type)")
plt.show()
```

Conclusions - Dining Restaurants have a majority of offline orders, whereas the cafes have more of the online orders.
